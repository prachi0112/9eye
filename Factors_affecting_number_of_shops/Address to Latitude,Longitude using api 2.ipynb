```python
for x in range(len(addresses)):
    try:
        payload1= {'address': addresses['ADDRESS2'][x] }
        url= "https://atlas.mapmyindia.com/api/places/geocode?"
        headers = {'Authorization': access_token} #Enter Your Access Code


        r1=requests.get(url,headers=headers, params=payload1)
        text_json = json.loads(r1.text)
        addresses['lat'][x] = text_json['copResults']['latitude']
        addresses['long'][x] = text_json['copResults']['longitude']
        addresses['district'][x] = text_json['copResults']['district']
    except IndexError:
        print("Address was wrong...")
    except Exception as e:
        print("Unexpected error occurred.", e )
addresses.head()
```
